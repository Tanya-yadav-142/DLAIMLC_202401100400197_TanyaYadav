```python
import tensorflow as tf
import numpy as np
```

The neural networkвЂ™s job is to learn this relationship by itself

```python
#step1:You are giving examples to the AI and vo apne aap indono k beech ka relation smj le
#dtype=float?
#Neural networks work with decimal numbers internally.
#smooth gradient calculation
# better learning
x=np.array([1,2,3,4],dtype=float)
y=np.array([2,4,6,8],dtype=float)

#step2: Build the neural network
#sequential mtlb go layer by layer
#input -> layer ->output
model=tf.keras.Sequential([
    #will take input 1 element at 1 time as input
    tf.keras.Input(shape=(1,)),
    #dense=1 fully connected
    #and it will create 1 neuron which will have exact formula inside it(output=wx+b)
    tf.keras.layers.Dense(1)
])

#step3
#вЂњHow should I learn?вЂќ y hum model ko batate h and
#SGD = Stochastic Gradient Descent it adjusts weight accordingly to reduce error
#mse loss=(realв€’predicted)2
model.compile(optimizer='sgd',loss='mean_squared_error')

#step4
#go through entire data 10 times
model.fit(x,y,epochs=10)
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

#convert lists to Numpy arrays
x = np.array([1, 2, 3, 4],dtype=float)
y = np.array([2,4,6,8],dtype=float)

#define model
model = Sequential([
    Dense(1, input_shape=(1,))
])

#compile model
model.compile(optimizer='sgd', loss='mean_squared_error')

#train model
model.fit(x,y,epochs=10)
```

```python
#in pytorch we have to do it manually
import torch
import torch.nn as nn
import torch.optim as optim

#define simple linear model
model=nn.Linear(1,1)

#loss and optimizer
criterion=nn.MSELoss()
optimizer=optim.SGD(model.parameters(),lr=0.01)

#dummy data
x=torch.tensor([[1.],[2.],[3.],[4.]])
y=torch.tensor([[2.],[4.],[6.],[8.]])

#raining loop
for epoch in range(10):
    #forward pass
    optimizer.zero_grad()
    outputs=model(x)
    loss=criterion(outputs,y)
    loss.backward()
    optimizer.step()
    print(f"Epoch {epoch+1}, Loss: {loss.item()}")
```
