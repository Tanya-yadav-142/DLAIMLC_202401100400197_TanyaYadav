#**BERT = Bidirectional Encoder Representations from Transformers**

##🔧 How it works internally(Sentiment analysis)

##How it works internally(Question answering)

🧠 ONE-LINE MASTER ANSWER

“This code uses Hugging Face pipelines with pre-trained BERT models to perform sentiment classification and extractive question answering without training from scratch.”

```python
# pip install transformers datasets torch
from transformers import pipeline

# ── Sentiment ──
sent = pipeline("sentiment-analysis", model="textattack/bert-base-uncased-SST-2")
print(sent(["I love this movie!", "This was terrible."]))

# ── Question Answering ──
qa = pipeline("question-answering", model="deepset/bert-base-cased-squad2")
context = "BERT was introduced by Google in 2018. It uses transformer encoders."
print(qa(question="Who made BERT?",  context=context))
print(qa(question="When was BERT made?", context=context))
```
